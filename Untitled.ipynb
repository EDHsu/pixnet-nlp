```python
# -*- coding: utf-8 -*-
%matplotlib inline
import codecs, re, jieba
import gensim
import difflib
import numpy as np
import random
from smart_open import smart_open
import matplotlib.pyplot as plt
jieba.add_word('龐傍燮謝', freq=10, tag='xx')

def test_neg(n = 10):
    a = np.zeros(vocab_size)
    b = np.zeros(vocab_size)
    for i in range(vocab_size):
        a[i] = np.dot(syn1neg[i], model.syn0[i])
        l = 0
        f = 0
        for j in range(n):
            k = random.choice(range(vocab_size))
            if k != i:
                f = f + np.dot(syn1neg[i], model.syn0[k])
                l = l + 1
        b[i] = f / l
    plt.figure(figsize=(24,12))
    plt.grid()
    plt.plot(a)
    plt.plot(b, 'r')
```

```python
path = 'w2v-experiment/model/v100_w5_n50_p10/'
model = gensim.models.Word2Vec.load_word2vec_format(path + 'syn0.bin', binary = True)
vocab_size, vector_size = model.syn0.shape
syn1neg = np.zeros((vocab_size, vector_size), dtype=np.float32)
binary_len = np.dtype(np.float32).itemsize * vector_size
with smart_open(path + 'syn1neg.bin') as fin:
    for i in range(vocab_size):
        weights = np.fromstring(fin.read(binary_len), dtype=np.float32)
#         if weights.sum() != 0:
#             syn1neg[i] = weights / ((weights * weights).sum() ** 0.5)
#         else:
        syn1neg[i] = weights
```

```python
def simple_preprocess(sample):
    no, content, a, b, c, d, e = re.findall(r'\[(\d+)\](.*)### a:(.*), b:(.*), c:(.*), d:(.*), e:(.*)\[END\]', sample)[0]
    options = [a.strip(), b.strip(), c.strip(), d.strip(), e.strip()]
    content = re.sub('(")(.?)', r'\2', content.strip())
    print(content)
    content = content.strip().replace('︽⊙＿⊙︽', '龐傍燮謝')
    wlist = list(jieba.cut(content))
    qidx = []
    i = 0
    for w in wlist:
        if w == '龐傍燮謝':
            wlist[i] = '*'
            qidx.append(i)
        i += 1
    return (no, wlist, qidx, options)

path = '../2016-pixnet-ai/pixbot/question_samples/'
# filename = '2016-07-31-15-02-02.txt'
# no = 4
filename = '2016-07-31-20-12-45.txt'
no = 5
def get_sample(file_name, no_sample):
    with codecs.open(file_name, 'r', 'utf-8') as f:
        for line in f:
            for (no, content, a, b, c, d, e) in re.findall(r'\[(\d+)\](.*)### a:(.*), b:(.*), c:(.*), d:(.*), e:(.*)\[END\]', line):
                if int(no) == int(no_sample):
                    return line

no, wlist, qidx, anslist = simple_preprocess(get_sample(path + filename, no))
print(no)
print(wlist)
print(qidx)
print(anslist)
```
