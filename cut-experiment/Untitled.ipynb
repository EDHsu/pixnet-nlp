```python
import json
import re
import jieba
from time import time
```

```python
my_dict = {}
t = time()
i, j = 0, 0
with open('../data/zhwiki-zh-clean', 'wb') as g:
    with open('../data/zhwiki-latest-all-titles-in-ns0', 'rb') as f:
        for line in f:
            j += 1
            s = line.decode('utf-8')[:-1]
            m = re.match(r'.*?[^\u4E00-\u9FFF]+.*?', s)
            if m: continue
    #         s = re.sub(r'\(.*\)', r'', s).strip()
            s = s.strip()
            if len(s) > 5: continue
            if s in my_dict: continue
            my_dict[s] = 0
            g.write((s + '\n').encode('utf-8'))
            i += 1
    #         print(s)
            if(i % 100000 == 1):
                print(i, j)
print(len(my_dict))
print(time() - t)
```
